Funciones lambda, una funcion a la cual no hay que defenirle un nombre y devuelve automaticamente el resultado (no se usa el return)

```python
suma = lambda a, b: a+b
suma(3, 4)
```

```python
multiplicar = lambda a, b: a*b
multiplicar(4, 3)
```

```python
ultimoel = lambda palabra: palabra[-1]
ultimoel("casa")
```

```python
list(map(lambda x: x**2, [1, 2, 3, 4]))
```

```python
par = lambda numero: numero % 2 == 0
par(5)
```

```python
list(filter(par, (1, 2, 3, 4)))
```
