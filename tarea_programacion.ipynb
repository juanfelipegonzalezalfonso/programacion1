```python
print("1. Desde una lista:")
lista_original = [10, 20, 30]
print("  Lista original:", lista_original)

# a tupla
tupla_desde_lista = tuple(lista_original)
print("  a tupla:", tupla_desde_lista)

# a conjunto
conjunto_desde_lista = set(lista_original)
print("  a conjunto:", conjunto_desde_lista)

# a diccionario
diccionario_desde_lista = dict(enumerate(lista_original))
print("  a diccionario (con enumerate):", diccionario_desde_lista)
```

```python
print("2. Desde una tupla:")
tupla_original = ("10", "20", "30")
print("  tupla original:", tupla_original)

# a lista
lista_desde_tupla = list(tupla_original)
print("  a lista:", lista_desde_tupla)

# a conjunto
conjunto_desde_tupla = set(tupla_original)
print("  a conjunto:", conjunto_desde_tupla)

# a diccionario
diccionario_desde_tupla = dict(enumerate(tupla_original))
print("  a diccionario (con enumerate):", diccionario_desde_tupla)
```

```python
print("3. Desde un diccionario:")
diccionario_original = {"pizza favorita": "Hawaiana", "precio": 8000}
print("  diccionario original:", diccionario_original)

# a lista
lista_desde_diccionario = list(diccionario_original)
lista_desde_diccionario1 = list(diccionario_original.values())
lista_desde_diccionario2 = list(diccionario_original.items())
print("  a lista:", lista_desde_diccionario, lista_desde_diccionario1, lista_desde_diccionario2)

# a tupla
tupla_desde_diccionario = tuple(diccionario_original)
tupla_desde_diccionario1 = tuple(diccionario_original.values())
tupla_desde_diccionario2 = tuple(diccionario_original.items())
print("  a tupla:", tupla_desde_diccionario, tupla_desde_diccionario1, tupla_desde_diccionario2)


# a conjunto
conjunto_desde_diccionario = set(diccionario_original)
conjunto_desde_diccionario1 = set(diccionario_original.values())
conjunto_desde_diccionario2 = set(diccionario_original.items())
print("  a conjunto:", conjunto_desde_diccionario, conjunto_desde_diccionario1, conjunto_desde_diccionario2)
```

```python
print("4. Desde un conjunto:")
conjunto_original = {"gato", "perro", "cangrejo"}
print("  conjunto original:", conjunto_original)

# a lista
lista_desde_conjunto = list(conjunto_original)
print("  a lista:", lista_desde_conjunto)

# a tupla
tupla_desde_conjunto = tuple(conjunto_original)
print("  a tupla:", tupla_desde_conjunto)


# a diccionario
diccionario_desde_conjunto = dict(enumerate(conjunto_original))
print("  a diccionario (con enumerate):", diccionario_desde_conjunto)
```
