# While : meintras que


```python
a=0
b=1
while b < 100:
  print(b)
  a = b
  b = a+b
```

```python
a=0
b=1
while b < 100:
  print(b)
  a, b = b, a+b
```

#Range:

```python
a, b = 0,1
for b in range(20):
  print(b)
  a, b = b, a + b
```

```python
x = int(input("ingresa un numero: "))
if x % 2 == 0:
  print("el numro es par.")
else:
  print("el numero es impar.")
```

```python
x = int(input("ingresa un numero: "))
if x % 3 == 0:
  print("el numro es multiplo de 3.")
elif x %3 ==1
  print("no es multiplo de 3 porque el residuo es 1")
else:
  print("el numero no es multiplo de 3.")
```
