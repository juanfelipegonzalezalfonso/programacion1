El metodo difference lee ambos conjuntos y devulve en un nuevo conjunto el elemento que diferencie al primero del segundo o en este caso a A de B

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  vocalesA = "AEIOU"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  consonantesA = "BCDFGHJKLMNÑPQRSTVWXYZ"
  numeros = "1234567890"
  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0
  contador5 = 0

  for letra in palabra:
    if letra in vocales:
      contador += 1
    elif letra in vocalesA:
      contador1 += 1
    elif letra in consonantes:
      contador2 += 1
    elif letra in consonantesA:
      contador3 += 1
    elif letra in numeros:
      contador4 += 1
    else:
      contador5 += 1
  return({"Cantidad de vocales minusculas:": contador,
"cantidad de vocales mayusuclas": contador1,
"Cantidad de consonates minusculas:": contador2,
"Cantidad de consonantes mayusculas": contador3,
"Cantidad de numeros": contador4,
"Cantidad de otros caracteres:": contador5})
```

```python
palabra = input("ingrese una palabra: ")
contador_vocales1(palabra)
```

Asigné diferentes clasificaciones a los caracteres posibles dentro de la palabra, con el for vamos recorriendo la palabra y comprobando si el elemento se encuentra dentro de algunas de las anteriores clasificaciones y se suma 1 al contador correspondiente a dicha categoría, si no está entre alguna de estas entonces corresponde a otro caracter y se le suma a ese contador, al final imprimimos el numero correspondiente a cada carecter con un texto que indique a cual pertenece
