```python
def sumar_matrices(A, B):
  filas = len(A)
  columnas = len(A[0])
  # Verificar que las dimensiones sean iguales
  if filas != len(B) or columnas != len(B[0]):
    return "Error: Las matrices deben tener las mismas dimensiones para sumarse."

  else:
    resultado = []
    for i in range(filas):
      fila = []
      for j in range(columnas):
        fila.append(A[i][j] + B[i][j])
      resultado.append(fila)
    return resultado
```

```python
A = [[1, 2], [3, 4]]
B = [[1, 2], [5, 6]]
sumar_matrices(A, B)
```

```python
def contador_vocales1(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  numeros = "0123456789"
  espacio = " "
  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    elif letra in numeros:
      contador2 += 1
    elif letra in espacio:
      contador3 += 1
    else:
      contador4 += 1
  return({"Cantidad de vocales:": contador,
          "Cantidad de consonates:": contador1,
          "Cantidad de números:": contador2,
          "Cantidad de espacios:": contador3,
          "Cantidad de otros caracteres:": contador4})
```
